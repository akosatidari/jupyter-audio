```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import librosa
import librosa.display
import IPython.display as ipd
import seaborn as sns
from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

Importujeme potrebné knižnice a nastavíme štýl zobrazovania grafov

```python
Fs = 48000
lenSec = 1
N = Fs*lenSec
```

Nastavíme frekvenciu, trvanie signálu a počet vzoriek

```python
noise = np.random.rand(N,1)
rohdB = -3
roh = pow(10,(rohdB/20))
```

Vygenerujeme pole N náhodných čísel, nastavíme úroveň hlasitosti šumu v decibeloch a v poslednom riadku prevedieme úroveň hlasitosti z decibelov na amplitúdu

```python
xPeak = max(abs(noise))
noiseN = (roh/xPeak) * noise
noiseN= noiseN.flatten()
```

Nájdeme maximálnu hodnotu amplitúdy, čím normalizujeme vytvorený biely šum na danú hodnotu hlasitosti rohdB, a nakoniec prevedieme biely šum z dvojrozmerného poľa na jednorozmerné, aby sme mohli pracovať s inými jednorozmernými poľami.

```python
pd.Series(noiseN).plot(figsize=(15,7),
                  lw=1,
                  title='Biely šum',
                  color=color_pal[0])
plt.show()
```

Tento kód vytvorí graf, ktorý zobrazuje generovaný biely šum s normalizovanou amplitúdou.
