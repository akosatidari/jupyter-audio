```python
import random
import numpy as np
import pandas as pd
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import colorednoise as cn
import seaborn as sns
from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

Importujeme potrebné knižnice a nastavíme štýl zobrazovania grafov

```python
beta = 1         
samples = 2**16
```

Beta je indikátorom mocninového zákona, ktorý sa používa na generovanie rôznych typov šumu: 0-biely, 1-ružový, 2-červený (Brownov). V premennej samples definujeme počet vzoriek, z ktorých sa bude šum generovať.

```python
A = cn.powerlaw_psd_gaussian(beta, samples)
```

Tento riadok generuje náhodný šum pomocou funkcie powerlaw_psd_gaussian z knižnice colourednoise. Funkcia prijíma dva argumenty: beta, ktorý predstavuje exponent mocninového zákona, a samples, ktorý predstavuje počet vygenerovaných náhodných hodnôt. Výsledok sa uloží do premennej A.

```python
pd.Series(A).plot(figsize=(15,7),
                  lw=1,
                  title='Ružový šum',
                  color=color_pal[0])
plt.show()
```

Graf zobrazuje náhodný šum generovaný funkciou cn.powerlaw_psd_gaussian
