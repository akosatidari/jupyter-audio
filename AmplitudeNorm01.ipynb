```python
# Peak amplitude normalization

import pandas as pd
import librosa
import matplotlib.pylab as plt
import seaborn as sns
from itertools import cycle
import numpy as np

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])

def dbAmpChange(input, dBChange):
    G =10** (dBChange/20) 
    output = G*input
    return output 

x, Fs = librosa.load("../holovachm/bakalarka/sounds02/piano.wav") # cesta na .wav file 

# Zvolime konstantu rohdB <=0 a vypocitame roh
rhodB = -1
rho = 10**(rhodB/20);

xPeak = max(abs(x));
yPeakNorm = (rho/xPeak) * x;

pd.Series(yPeakNorm).plot(figsize=(15,7),
                  lw=1,
                  title='Raw Audio Example',
                  color=color_pal[0])
pd.Series(x).plot(figsize=(15,7),
                  lw=1,
                  title='Raw Audio Example',
                  color=color_pal[1])
plt.show()
```
